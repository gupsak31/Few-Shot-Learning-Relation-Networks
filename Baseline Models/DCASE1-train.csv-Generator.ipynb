```python
import numpy as np
import pandas as pd
import random
```

```python
df = pd.read_csv("dcase1.csv")
```

```python
df["class"].value_counts()
```

```python
def fewshot_class(class_name):
    df_class = df.loc[df["class"] == class_name]
    indexes = [i for i in range(df_class.shape[0])]
    random.shuffle(indexes)
    df_class_train = df_class.iloc[indexes[:50]]
    df_class_test = df_class.iloc[indexes[50:]]
    return df_class_train, df_class_test
```

```python
def sample_class(class_name):
    df_class = df.loc[df["class"] == class_name]
    indexes = [i for i in range(df_class.shape[0])]
    random.shuffle(indexes)
    df_class_train = df_class.iloc[indexes[:1000]]
    return df_class_train
```

```python
df_airport_train, df_airport_test = fewshot_class("airport")
df_publicsquare_train, df_publicsquare_test = fewshot_class("public_square")
df_tram_train, df_tram_test = fewshot_class("tram")
```

```python
df_shoppingmall = sample_class("shopping_mall")
df_streetpedestrian = sample_class("street_pedestrian")
df_metro = sample_class("metro")
df_park = sample_class("park")
df_metrostation = sample_class("metro_station")
df_bus = sample_class("bus")
df_streettraffic = sample_class("street_traffic")
```

```python
df_fewshot_train = pd.concat([df_shoppingmall, df_streetpedestrian, df_metro, df_park, df_metrostation, df_bus, df_streettraffic, df_airport_train, df_publicsquare_train, df_tram_train], ignore_index = True)
df_fewshot_train["class"].value_counts()
```

```python
df_fewshot_test = pd.concat([df_airport_test, df_publicsquare_test, df_tram_test], ignore_index = True)
df_fewshot_test["class"].value_counts()
```

```python
df_fewshot_train.to_csv("dcase1_train.csv", index = False)
df_fewshot_test.to_csv("dcase1_other.csv", index = False)
```
