```python
import numpy as np
import pandas as pd
import random
```

```python
df = pd.read_csv("UrbanSound8K.csv")
```

```python
def fewshot_class(class_name):
    df_class = df.loc[df["class"] == class_name]
    indexes = [i for i in range(df_class.shape[0])]
    random.shuffle(indexes)
    df_class_train = df_class.iloc[indexes[:20]]
    df_class_test = df_class.iloc[indexes[20:]]
    return df_class_train, df_class_test

def train_class(class_name):
    df_class = df.loc[df["class"] == class_name]
    indexes = [i for i in range(df_class.shape[0])]
    random.shuffle(indexes)
    df_class_train = df_class.iloc[indexes[:500]]
    df_class_test = df_class.iloc[indexes[500:]]
    return df_class_train, df_class_test
```
